```python
# File: notebooks/eda.ipynb

import pandas as pd
import os

# Paths
raw_path = '../data/raw/redfin_master_data.tsv000'
processed_path = '../data/processed/'

# Load TSV file
df = pd.read_csv(raw_path, sep='\t')

# Preview
print("Full dataset shape:", df.shape)
print(df.columns.tolist())

# Step 1: Filter to San Diego Metro
df_sd = df[df['REGION'] == 'San Diego, CA metro area'].copy()

# Step 2: Select columns we care about
columns_of_interest = [
    'PERIOD_BEGIN',
    'MEDIAN_SALE_PRICE',
    'INVENTORY',
    'HOMES_SOLD'
]
df_sd = df_sd[columns_of_interest]

# Step 3: Clean + rename columns
df_sd['PERIOD_BEGIN'] = pd.to_datetime(df_sd['PERIOD_BEGIN'])
df_sd = df_sd.rename(columns={
    'PERIOD_BEGIN': 'date',
    'MEDIAN_SALE_PRICE': 'median_price',
    'INVENTORY': 'inventory',
    'HOMES_SOLD': 'homes_sold'
})

# Step 4: Sort by date
df_sd = df_sd.sort_values('date')

# Step 5: Save cleaned data
os.makedirs(processed_path, exist_ok=True)
output_file = os.path.join(processed_path, 'san_diego_housing_metrics.csv')
df_sd.to_csv(output_file, index=False)

print("✅ Cleaned data saved to:", output_file)
df_sd.tail()
```
